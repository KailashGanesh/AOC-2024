## Actual solve

```python
def check(one_array):
    incrasing = all([num1 < num2 for num1, num2 in zip(one_array, one_array[1:])])

    decre = all([num1 > num2 for num1, num2 in zip(one_array, one_array[1:])])

    diff_by = all([abs(num1 - num2) >= 1 and abs(num1 - num2) <= 3 for num1, num2 in zip(one_array, one_array[1:]) ])
    # print(incrasing,  decre,  diff_by)
    if (incrasing or decre) and diff_by:
        # print(one_array, "safe")
        return True
    else:
        # print(one_array, "unsafe")
        return False
```

```python
# part 1
for i in diff_arr:
    print(check(i))
```

```python
# part 2
count = 0
safe = False

for i in diff_arr:
    print("array being used",i)
    safe = False
    for j in range(len(i)):
        if safe:
            continue
        dup = i.copy()
        dup.pop(j)
        if check(dup):
            print("safe")
            count += 1
            safe = True

    print("unsafe")
        


print(count)
```
