```python
import re

regex = re.compile(r"mul\(\d+,\d+\)")
muls = []
for line in data:
    muls.append(re.findall(regex,line))
    print(muls)

muls_flattend = [mul for row in muls for mul in row]

import numpy as np
output = 0

for mul in muls_flattend:
    two_nums = np.array(mul.replace("mul(", "").replace(")", "").split(",")).astype(int)

    output += np.prod(two_nums)

print(output)    
```

```python
# part 2

regex_p2 = re.compile(r"don't\(\)|do\(\)|mul\(\d+,\d+\)")

acc = []
for line in data:
    acc.append(re.findall(regex_p2, line))

acc_flattend = [mul for row in acc for mul in row]
acc_flattend
```

```python
do = True
do_muls = []

for i in acc_flattend:
    if "don't()" == i: 
        do = False
    if "do()" == i:
        do = True
    
    if do and "do()" != i:
       do_muls.append(i) 
```
